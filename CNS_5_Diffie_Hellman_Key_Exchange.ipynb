

# DIFFIE HELLMAN KEY EXCHANGE ALGORITHM: Using Man-in-middle Attack



*   	Alice and Bob agree on a large prime numbers n  and another number g such that g<n and g is a   
primitive root of n

*   	Attacker Tom is listening. So Tom is also aware of  n and g
* 	Alice and Bob select random numbers x and y. Tom also selects his own x and y
* 	Alice calculates A,  Bob calculates B,  Tom calculates his own A and B
* Alice transmits A to Bob, which is intercepted by Tom. Tom, replaces Alices A  with his own A and forwards it to Bob.
* Similarly, Tom intercepts Bobs B and forwards his own B to Alice.
* Now Alice finds K_1, Bob finds K_2, however, Tom finds both K_1 and K_2




```python
import random
n=11
g=7

Alice_x=random.randint(1, n-1)
Bob_y=random.randint(1, n-1)
Z_Tom_x=random.randint(1, n-1)
Z_Tom_y=random.randint(1, n-1)

Alice_secret_A= (g**Alice_x)%n
Bob_secret_B= (g**Bob_y)%n

Tom_secret_A= (g**Z_Tom_x)%n
Tom_secret_B= (g**Z_Tom_y)%n

K1_Alice= ((g**Z_Tom_y)**Alice_x )% n
K2_Bob= ((g**Z_Tom_x)**Bob_y )% n

K1_Tom=((g**Alice_x)**Z_Tom_y )% n
K2_Tom=((g**Bob_y)**Z_Tom_x )% n

print("Key gerated by Alice (K1): ",K1_Alice )
print("Key gerated by Bob (K2): ",K2_Bob )

print("Key gerated by Tom (K1): ",K1_Tom )
print("Key gerated by Tom (K2): ",K2_Tom )
```
